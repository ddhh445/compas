```r
#install.packages('CORElearn')
library(CORElearn)
```

```r
data=read.csv('relif_data.csv')
```

```r
df=data[4:23]
```

```r
#인도여부는 나중에..
df=df[,-c(19)]
```

```r
RE=attrEval(target~.,data=df,estimator='Relief',ReliefIterations=30)
```

```r
SRE=sort(RE,decreasing=T)
```

```r
plot(1:length(SRE),SRE,type='b',ylab='Separability',
    xlab='variable',main='Variable importance plot',xaxt='n')
axis(side=1,at=1:length(SRE),labels=names(SRE),cex.axis=0.8,las=2)
```

```r
#separabilty 가 낮은건 imblance 하기 때문인듯 한다 치더라도 
#넓게 잡아도 대략 8~10개 정도 변수만 쓸 수 있을 듯 하다
```

```r
#RElifF 알고리즘
REF=attrEval(target~.,data=df,estimator='ReliefFequalK',
            ReliefIterations=30)
```

```r
SREF=sort(REF,decreasing = T)
```
